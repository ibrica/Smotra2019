```python
def download_hotdog_images(trainOrTest, coco):
    cats = coco.loadCats(coco.getCatIds())
    nms=[cat['name'] for cat in cats]
    # print('COCO categories: \n{}\n'.format(' '.join(nms)))
    catIds = coco.getCatIds(catNms=['hot dog'])
    imgIds = coco.getImgIds(catIds=catIds )
    images = coco.loadImgs(imgIds)
    #print("imgIds: ", imgIds)
    print(trainOrTest + " images nr: ", len(images))
    directory = "data/" + trainOrTest +  "/hotdog/"
    for im in images:
        # print("im: ", im)
        img_data = requests.get(im['coco_url']).content
        with open(directory + im['file_name'], 'wb') as handler:
            handler.write(img_data)
```

```python
def download_nothotdog_images(trainOrTest, coco):
    cats = coco.loadCats(coco.getCatIds())
    nms=[cat['name'] for cat in cats]
    # print('COCO categories: \n{}\n'.format(' '.join(nms)))
    catIds = coco.getCatIds(catNms=['pet','banana'])
    imgIds = coco.getImgIds(catIds=catIds )
    images = coco.loadImgs(imgIds)
    #print("imgIds: ", imgIds)
    print(trainOrTest + " images nr: ", len(images))
    limit = 1222 if trainOrTest == 'train' else 51
    directory = "data/" + trainOrTest +  "/nothotdog/"
    count = 0
    for im in images:
        count += 1
        if count > limit:
            break
        print("Spremam: " + im['file_name'])
        img_data = requests.get(im['coco_url']).content
        with open(directory + im['file_name'], 'wb') as handler:
            handler.write(img_data)
```
